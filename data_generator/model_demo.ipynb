```python
import numpy as np
import matplotlib.pyplot as plt

def integrate_neuron(u_i, v_i, iapp, DT):
    u_new = u_i + DT * 10*((1)*(u_i*(u_i-0.4)*(1-u_i))-v_i  +iapp );
    v_new = v_i + DT * ((u_i*0.04-0.16*v_i));
    return u_new, v_new

# Fixed pairs of initial conditions for u and v
e=.0
u_initials =  np.ones(3)*0.3875185889005661*np.random.uniform(1-e,1+e,size=(3))                         
v_initials = np.ones(3)*0.09100821733474732  *np.random.uniform(1-e,1+e,size=(3))  
# Fixed iapp values
iapp_values = np.ones(3)*0.09   *np.random.uniform(1-e,1+e,size=(3))  

DT = 0.01
num_steps = int(50/DT)

# Plotting
fig, axs = plt.subplots(len(u_initials), len(v_initials), figsize=(12, 8))

for i,x0 in enumerate(zip(u_initials,v_initials)):

        u_values = np.zeros(num_steps)
        v_values = np.zeros(num_steps)
        # Perform integration

        for j, iapp in enumerate(iapp_values):
            u0,v0=x0
            u,v=x0
            print(u,v,iapp)

            for t in range(num_steps):
                u_values[t] = u
                v_values[t] = v
                u, v = integrate_neuron(u, v, iapp, DT)
            # Plotting
            # Choosing color based on iapp index
            axs[i, j].plot(u_values, label="U")
            axs[i, j].plot(v_values, label="V")
            
            axs[i, j].set_xlabel('Time Step')
            axs[i, j].set_ylabel('Value')
            axs[i, j].set_title(f'u0={u0}, v0={v0}, iapp={iapp}')
            axs[i, j].legend()
 
plt.tight_layout()
plt.savefig("f1.pdf")
plt.show()
```
